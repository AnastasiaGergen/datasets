```python
import pandas as pd
import numpy as np
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
df_sea = pd.read_csv('../input/charleston-sea-level-daily-dataset-1921-2014/daily Charleston 1921-2014.csv')
df_sea
```

```python
df_sea.drop([0], inplace=True)
df_sea['date'] = pd.to_datetime(df_sea['time']).dt.to_period('D')
df_sea.index = df_sea['date']
df_sea.drop(columns=['depth', 'latitude', 'longitude', 'sensor_type_code', 'date', 'time'], inplace=True)
df_sea.columns = ['height']
df_sea['height'] = pd.to_numeric(df_sea['height'])
```

```python
df_sea.isna().sum() / len(df_sea) * 100 # % значений пропущено
```

```python
df_sea['height'] = df_sea['height'].fillna(method='bfill')
df_sea
```

```python
df_prod = pd.read_csv('../input/industrialproductionelectricandgasutilities/IPG2211A2N.csv')
df_prod
```

```python
df_prod.columns = ['date', 'production']
df_prod.index = pd.to_datetime(df_prod['date'])
df_prod.drop('date', axis=1, inplace=True)
df_prod
```

```python
df_sea.to_csv('sea_height.csv', index=False)
df_prod.to_csv('industrial_production.csv', index=False)
```
